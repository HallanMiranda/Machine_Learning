```python
import numpy as np

from sklearn          import datasets as ds
from sklearn          import cluster  as ct
from sklearn          import metrics  as mt 
from matplotlib       import pyplot   as plt
from sklearn.datasets import make_moons
```

# 1.0 Load Data

n_samples: É o número total de pontos de dados a serem gerados. Neste caso, você está criando 100 pontos de dados.
n_features: Indica a dimensão do espaço de características, ou seja, o número de atributos em cada ponto de dados. Aqui, você está gerando pontos de dados com 2 atributos.
centers: Refere-se ao número de clusters ou grupos que você deseja criar. Neste caso, você está definindo 3 clusters, o que significa que os dados serão gerados em torno de 3 centros distintos.
cluster_std: Este parâmetro controla o desvio padrão dos clusters. Quanto menor for o valor, mais compactos serão os clusters. Valores maiores resultarão em clusters mais dispersos.
random_state: É uma semente para a geração de números aleatórios. Definir esse valor permite que você reproduza os mesmos resultados sempre que executar o código com o mesmo valor.

```python
# X, y = ds.make_blobs(n_samples=100,
#                      n_features=4,
#                      centers=3,
#                      cluster_std= 0.3,
#                      random_state=0)

X, y = make_moons(n_samples=500, noise=0.1)

plt.scatter( X[:, 0], X[:, 1], c=y)
```

```python
k= 2

# Teste impirico , ao inves de determinar um unioc numero, eu testo uma lista de numeros
clusters = np.arange(2, 21, 1)# vl inicial , vl final , onde o incrementoeu deido e q 1 e vai variar de 1 a 1  

ss_list = []

for k in clusters:
    # Modelo Define
    model = ct.KMeans( n_clusters=k, init= 'random', n_init=10, random_state=0)
    
    # Fit (treinamento)
    labels = model.fit_predict( X )

    # Performace
    ss = mt.silhouette_score( X, labels )# encontra rrelacao entre distancia e coesaion 
    ss_list.append( ss )

    print('Cluster X: {} - ss: {})'.format( k, ss))
```

```python
plt.plot( clusters, ss_list, marker='o', linewidth=2 );
plt.xlabel( 'K Values');
plt.ylabel( 'Silhouette Score Avg');
```
